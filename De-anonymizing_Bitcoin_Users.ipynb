# Append all addresses for a transaction

```python
def get_addr_for_transaction(data):
    transactions = dict()
    for index, row in data.iterrows():
        if index%1000 == 0:
            print(index)
        addr = row['addrID']
        trans_id = row['txID']
        if trans_id in transactions:
            prev_addrs = transactions[trans_id]
            prev_addrs.add(addr)
            transactions[trans_id] = prev_addrs
        else:
            new_addrs = set()
            new_addrs.add(addr)
            transactions[trans_id] = new_addrs
    return transactions
    
```

# Get the amount of incoming bitcoin and outgoing bitcoin for transactions

```python
def get_bitcoin_sum_for_transaction(data):
    transactions_sum = dict()
    for index, row in data.iterrows():
        if index%1000 == 0:
            print(index)
        addr = row['addrID']
        trans_id = row['txID']
        bit_value = row['sum']
        if trans_id in transactions_sum:
            prev_addrs = transactions_sum[trans_id]
            addr_sum = (addr, bit_value)
            prev_addrs.add(addr_sum)
            transactions_sum[trans_id] = prev_addrs
        else:
            new_addrs = set()
            addr_sum = (addr, bit_value)
            new_addrs.add(addr_sum)
            transactions_sum[trans_id] = new_addrs
    return transactions_sum
```

# Identify the change address using heuristic 2

```python
def find_change_address(in_transaction, out_transaction):
    change_addr_list = dict()
    for trans, addr_arr in in_transaction.items():
        if trans in out_transaction:
            out_addr_arr = out_transaction[trans]
            b_found = False
            change_addr = ""
            if len(addr_arr)>1 and len(out_addr_arr)>1:
                for addr_sum_pair in out_addr_arr:
                    out_addr = addr_sum_pair[0]
                    bit_sum = addr_sum_pair[1]
                    for in_addr_sum_pair in addr_arr:
                        if bit_sum < in_addr_sum_pair[1]:
                            change_addr = out_addr
                            b_found = True
                            break
                    if b_found == True:
                        break
            if b_found == True:
                input_addr = set()
                for in_addr in addr_arr:
                    input_addr.add(in_addr[0])
                change_addr_list[change_addr] = input_addr
    
    return change_addr_list
```

# Creating the graph following Heuristic 1

```python
def create_graph_heuristic_1(transactions, in_graph):
    for key, addr in transactions.items():
        value = list(addr)
        for i in range(len(value)-1):
            if value[i] in in_graph:
                existing_list = in_graph[value[i]]
                existing_list.append(value[i+1])
                in_graph[value[i]] = existing_list
            else:
                new_list = []
                new_list.append(value[i+1])
                in_graph[value[i]] = new_list

            if value[i+1] in in_graph:
                existing_list = in_graph[value[i+1]]
                existing_list.append(value[i])
                in_graph[value[i+1]] = existing_list
            else:
                new_list = []
                new_list.append(value[i])
                in_graph[value[i+1]] = new_list

        if len(value) == 1:
            if value[0] not in in_graph:
                in_graph[value[0]] = []
    return in_graph
```

```python
def update_graph_heuristic2(in_change_addr, in_graph):
    for chng_addr, in_addrs in in_change_addr.items():
        for addr in in_addrs:
            if chng_addr in in_graph:
                existing_list = in_graph[chng_addr]
                existing_list.append(addr)
                in_graph[chng_addr] = existing_list
            else:
                new_list = []
                new_list.append(addr)
                in_graph[chng_addr] = new_list

            if addr in in_graph:
                existing_list = in_graph[addr]
                existing_list.append(chng_addr)
                in_graph[addr] = existing_list
            else:
                new_list = []
                new_list.append(chng_addr)
                in_graph[addr] = new_list

    return in_graph
    
```

# Convert the vertices to id

```python
id = 0
for key, vlaue in graph.items():
    vertices_id_to_addr[id] = key
    addr_to_vertices_id[key] = id
    id +=1
```

# Apply Strongly Connected Component algorithm to find out all the adresses belong to same user.

```python
def DFSUtil(v, visited, c_components):
    visited[v] = 1
    c_components.append(v)
    v_addr = vertices_id_to_addr[v]
    adj_list = graph[v_addr]
    for adj_ver in adj_list:
        v_id = addr_to_vertices_id[adj_ver]
        if visited[v_id] == 0:
            DFSUtil(v_id, visited, c_components)
    
```

```python
def connected_component():
    connected_components = []
    visited = np.zeros(no_of_vertices)
    for v in range(no_of_vertices):
        if visited[v] == 0:
            c_components = []
            DFSUtil(v, visited, c_components)
            connected_components.append(c_components)
    return connected_components
```
